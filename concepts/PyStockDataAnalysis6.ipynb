```python
import pandas as pd
import datetime as dt
import pandas_datareader as web
import bs4 as bs
import pickle
import urllib2
import os
```

```python
def save_sp500_tickers():
    resp = urllib2.urlopen('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies').read()
    soup = bs.BeautifulSoup(resp, 'lxml')
    table = soup.find('table',{'class':'wikitable sortable'})
    tickers = []
    
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        ticker = str(ticker)
        tickers.append(ticker)
        
    with open('Data/sp500tickers.pickle','wb') as f:
        pickle.dump(tickers, f)
    return tickers
```

```python
def get_stock_data_from_yahoo(reload_sp500=False):
    
    tickers = []
    if reload_sp500:
        tickers = save_sp500_tickers()
    
    else:
        with open('Data/sp500tickers.pickle','rb') as f:
            tickers = pickle.load(f)
            
    if not os.path.exists('Data/S&P500'):
        os.makedirs('Data/S&P500')
    
    start = dt.datetime(2000, 1, 1)
    end =   dt.datetime(2017, 8, 31)
    
    for ticker in tickers:
        print ticker
        if ticker == "BRK.B" or ticker == "BF.B":
            continue
        if not os.path.exists('Data/S&P500/{}.csv'.format(ticker)):
            df = web.DataReader(ticker, 'yahoo', start, end)
            df.to_csv('Data/S&P500/{}.csv'.format(ticker))
        else:
            print "Ticker {} already present".format(ticker)
```
